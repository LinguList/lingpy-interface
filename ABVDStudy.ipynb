```python
# params: model choice, file extension
def run_workflow(mod, str_name):
    infile = "input/combined_fijian_maori.csv"
    flow = workflow.Workflow(infile)
    
    outfile_name = "fijian_maori_"+str_name+".csv"
    flow.cognate_detection(check=True, export="tsv", outfile=outfile_name)
    flow.score()
    return flow
```
