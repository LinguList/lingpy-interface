```python
# params: model, file extension
def run_workflow(chosen_model, file_ext):
    infile = "input/slavic.tsv"
    flow = interface.workflow.Workflow(infile)
    flow_tsv = interface.workflow.Workflow(infile)
    outfile_name = "slavic_results" + file_ext
    flow.cognate_detection(check=True, model=chosen_model, outfile=outfile_name)
    flow_tsv.cognate_detection(check=True, model=chosen_model, export="tsv", outfile=outfile_name)
    flow.score()
    return flow
```
